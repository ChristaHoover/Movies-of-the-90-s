```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
from time import sleep
from random import randint
```

```python
headers = {"Accespt-Language": "en-US,en;q=0.5"}
```

```python
pages = np.arange(1,150,50)
```

```python
movie_title = []
release_year = []
director = []
#lead_actor = []
#supporting_actor = []
#supporting_actor_two = []
#supporting_actor_three = []
movie_length = []
genre = []
film_rating = []
movie_gross = []
imdb_votes = []
metascore = []

for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?year=1990-01-01,1990-12-31&sort=boxoffice_gross_us,desc&start="+str(page)+"&ref_=adv_nxt")
    soup =bs(page.text, 'html.parser')
    #the top div that all the information is under
    movie_information = soup.findAll('div', attrs= {'class' : 'lister-item mode-advanced'})  
    
    # using sleep command
    sleep(randint(2,8))
    
    #Using another for loop to extract data from imdb
    for x in movie_information:
        
        
            #movie_title
        movie_name = x.h3.a.text
        movie_title.append(movie_name)

    
            #release_year
        movie_year = x.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
        release_year.append(movie_year)

          #director of the film
        movie_director = x.find('p', class_ = '').find_all('a')[0].text
        director.append(movie_director)

            #lead role
        #starring_actor = (x.find('p', attrs = {'class' : ''}).find_all('a')[1].text 
                     #if x.find('p', attrs = {'class' : ''}).find_all('a')[1] else 'none')
        #lead_actor.append(starring_actor)

            #supporting actor
        #actor_two = (x.find('p', attrs = {'class' : ''}).find_all('a')[2].text 
                     #if x.find('p', attrs = {'class' : ''}).find_all('a')[2] else 'none')
        #supporting_actor.append(actor_two)

            #supporting actor 2
        #actor_three = (x.find('p', attrs = {'class' : ''}).find_all('a')[3].text 
         #              if x.find('p', attrs = {'class' : ''}).find_all('a')[3] else 'none')
        #supporting_actor_two.append(actor_three)

            #supporting actor 3
        #actor_four =(x.find('p', attrs = {'class' : ''}).find_all('a')[4].text 
         #              if x.find('p', attrs = {'class' : ''}).find_all('a')[4] else 'none')
        #supporting_actor_three.append(actor_four)

            #finding the length of the movie and dropping the min so i only have a number
        movie_time = x.p.find('span', class_ = 'runtime').text.replace(' min', '')
        movie_length.append(movie_time)

            #finding the genre
        film_genre = x.find('span', class_ ='genre').text.replace('\n','').strip()
        genre.append(film_genre)

            #film rating Not Rated-R
        rating_for_film = x.find('span', class_ ='certificate')
        film_rating.append(rating_for_film)

            #imdb_votes and gross

        votes_and_gross = x.findAll('span', attrs = { 'name' : 'nv'})

            #gross
        gross = votes_and_gross[1].text.replace('M','')
        movie_gross.append(gross)

            #votes from imdb
        votes = votes_and_gross[0].text
        imdb_votes.append(votes)

            #imdb meta score
        meta = (x.find('span', attrs = {'class' : 'metascore'}).text.strip() 
                if x.find('span',  attrs = {'class' : 'metascore'}) else 'none')
        metascore.append(meta)
```

```python
imdb_list_df = pd.DataFrame()
imdb_list_df['movie_title'] = movie_title
imdb_list_df['release_year'] = release_year
imdb_list_df['director'] = director
#imdb_list_df['lead_actor'] = lead_actor
#imdb_list_df['supporting_actor'] = supporting_actor
imdb_list_df['movie_length'] = movie_length
imdb_list_df['genre'] = genre
imdb_list_df['film_rating'] = film_rating
imdb_list_df['movie_gross_in_millions'] = movie_gross
imdb_list_df['imdb_votes'] = imdb_votes
imdb_list_df['metascore'] = metascore
```
