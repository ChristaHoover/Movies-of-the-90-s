### finding the top 10 grossing movies in each year

```python
imdb_90 = imdb_wiki.loc[imdb_wiki.release_year == 1990].sort_values(by=['imdb_gross'],ascending=False).head(10)
imdb_90
```

```python
imdb_96 = imdb_wiki.loc[imdb_wiki.release_year == 1996].sort_values(by=['imdb_gross'],ascending=False).head(10)
imdb_96
```

```python
imdb_98 = imdb_wiki.loc[imdb_wiki.release_year == 1998].sort_values(by=['imdb_gross'],ascending=False).head(10)
imdb_98 
```

```python
imdb_wiki.sort_values(by = ['release_year', 'imdb_gross'], axis=0, ascending=[True, False], inplace=False,
               kind='quicksort', na_position='first', ignore_index=True, key=None)
```

```python
# concating the dfs so I have one that has all top ten grossing movies for each year making 100 movies
top_gross_movies= pd.concat([imdb_90,imdb_91,imdb_92,imdb_93,imdb_94,imdb_95,imdb_96,imdb_97,imdb_98,imdb_99], axis=0).reset_index()
top_gross_movies
```
