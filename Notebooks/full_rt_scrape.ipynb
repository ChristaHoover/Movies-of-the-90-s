```python
movie_title = []
tomato_score = []

for index,row in title_year.iterrows():
    movie=row.movie_title
    year = row.release_year
    test_movie = movie
    movie_title.append(movie)
    percentage = 'none'
    test_movie = test_movie.lower()
    
    if test_movie.find(' ') != -1:
        test_movie = test_movie.replace(' ', '%20')
    if test_movie.find(':') != -1:
        test_movie = test_movie.replace(':', '%3A')
    if test_movie.find("'") != -1:
        test_movie = test_movie.replace(':', '%60')  
    if test_movie.find('/') != -1:
        test_movie = test_movie.replace(':', '%2F') 
    sleep(random.randint(3,6))
    url = f"https://www.rottentomatoes.com/search?search={test_movie}"
    tomato_url = requests.get(url)
    #tomato_url.text
    soup = bs(tomato_url.text)
    movie_block = soup.find('search-page-result', attrs = {'type' : 'movie'})
   # movie_block = bs(movie_block)
    movie_options= movie_block.find('ul').findAll('search-page-media-row')#[0].get('tomatometerscore')
    for option in movie_options:
        if option.get('releaseyear') == str(year):
            percentage = option.get('tomatometerscore')
            tomato_score.append(percentage)
            print(f'found percentage {percentage} four {movie}')
    
test_df = pd.DataFrame({'movie_title': movie_title, 'tomato_score': tomato_score})
```
