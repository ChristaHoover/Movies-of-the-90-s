```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
from time import sleep
from random import randint
```

```python
#bring in url
url = 'https://editorial.rottentomatoes.com/guide/140-favorite-90s-movies/'
response = requests.get(url)
```

```python
response.text
```

```python
#making it more readable with prittify
soup = bs(response.content, 'lxml')
```

```python
tomato_info = soup.find_all('div', attrs = {'class' : 'col-sm-18'})
```

```python
movie_title = []
release_year = []
tomato_score = []
lead_role = []
actor_two = []
actor_three = []
actor_four = []
director = []
critics_consensus = []
synopsis= []
count_down = []

for x in tomato_info:
    
    #finding movie titles
    movie = x.h2.a.text
    movie_title.append(movie)
    
    #movie release year
    year = x.find('span', attrs = {'class' : 'subtle'}).text.replace('(','').replace(')','')
    release_year.append(year)
    
    #meter score in percent. 
    score = x.find('span', attrs = {'class' : 'tMeterScore'}).text.replace('%','')
    tomato_score.append(score)
    
    #Finding the cast
    info = x.findAll('a', attrs = {'class' : ''})
    
    #Lead role
    star =info[2].text
    lead_role.append(star)
    
    #second role
    sec =info[3].text
    actor_two.append(sec)
    
    #third role
    third = info[4].text
    actor_three.append(third)
    
    #fourth role
    fourth = info[5].text
    actor_four.append(fourth)
    
    #director
    direct = info[6].text
    director.append(direct)
    
    #critics consensus
    #try to figure out how to remove 'critics consensus'
    critics = x.find('div', attrs = {'class' : 'info critics-consensus'}).text.replace('Critics Consensus: ','')
    critics_consensus.append(critics)
    
    #Synopsis
    #try to figure out how to remove 'synopsis'
    synop = x.find('div', attrs = {'class' : 'info synopsis'}).text.replace('Synopsis: ','')
    synopsis.append(synop)
    
    #the number where the movie landed on the count down. 
    count = x.find('div', attrs = {'class' : 'countdown-index'}).text.replace('#','')
    count_down.append(count)
    
    #turning it into a dataframe
tomatos_list_df = pd.DataFrame()
tomatos_list_df['count_down'] = count_down
tomatos_list_df['movie_title'] = movie_title
tomatos_list_df['release_year'] = release_year
tomatos_list_df['tomato_score'] = tomato_score
tomatos_list_df['lead_role'] = lead_role
tomatos_list_df['actor_two'] = actor_two
tomatos_list_df['actor_three'] = actor_three
tomatos_list_df['actor_four'] = actor_four
tomatos_list_df['director'] = director
tomatos_list_df['critics_consensus'] = critics_consensus
tomatos_list_df['synopsis'] = synopsis
```

```python
tomatos_list_df.to_csv('tomato_list.csv')
```
